# Prepare Milgram

```python
import pandas as pd
import pathlib

RANDOM_SEED = 42
```

```python
name_list: pathlib.Path = pathlib.Path(
    "../../data/external/surnames.json"
)
path_to_participant_list: pathlib.Path = pathlib.Path(
    "../../data/prompt-fills/milgram/df_names.csv"
)
```

```python
# prepare dataframe of names
df_names = pd.read_json(name_list)
df_names = df_names.rename(columns={
    "Black or African American": "Black",
    "White": "White",
    "Asian and Native Hawaiian and Other Pacific Islander": "Asian",
    "American Indian and Alaska Native": "Indigenous",
    "Hispanic or Latino": "Latino"
})

# only keep top 10 most common names
df_names = df_names.head(10)

# reshape dataframe so that rows are ["Race", "Surname"]
df_names = df_names.melt()
df_names["Surname"] = df_names.apply(lambda x: x["value"].capitalize(), axis=1)
df_names = df_names.drop(columns=["value"])
df_names = df_names.rename(columns={"variable": "Race"})

# uncomment to save dataframe
df_names.to_csv(path_to_participant_list)

df_names.head()
```
